```python
import numpy as np
import matplotlib.pyplot as plt
from fastdtw import fastdtw

# 
np.random.seed(42)
trajectory_1 = np.cumsum(np.random.randn(10, 2), axis=0)  # 
trajectory_2 = np.cumsum(np.random.randn(10, 2), axis=0)  # 

# 
euclidean_distance = np.linalg.norm(trajectory_1[-1] - trajectory_2[-1])

# DTW
dtw_distance, _ = fastdtw(trajectory_1, trajectory_2)

# 
plt.figure(figsize=(10, 5))
plt.plot(trajectory_1[:, 0], trajectory_1[:, 1], label='Trajectory 1', marker='o')
plt.plot(trajectory_2[:, 0], trajectory_2[:, 1], label='Trajectory 2', marker='x')

# 
plt.title(f'Euclidean Distance: {euclidean_distance:.2f}, DTW Distance: {dtw_distance:.2f}')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
```
